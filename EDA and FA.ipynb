```python
country = pd.read_excel('Country-Code.xlsx')
```

```python
country.head()
```

```python
df.columns
```

```python
final = pd.merge(df,country,on='Country Code',how='left')
```

```python
final.head()
```

```python
##To check Data Types
final.dtypes
```

```python
country_name = final.Country.value_counts().index
```

```python
country_val = final.Country.value_counts().values
```

```python
## pie chart
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observation:Zomato maximum records or transaction are from India After that USA  and Then United Kingdoms

```python
final.columns
```

```python
final.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings = final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
- When Rating is between 4.5 to 4.9 --> Excellent
- When Rating are between 4.0 to 3.4 --> very good
- When Rating is between 3.5 to 3.9-->good
- When Rating is between 2.5 to 2.9 -->average
- When Rating is between 2.0 to 2.4 -->Poor

```python
ratings.head()
```

```python
import matplotlib.pyplot
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color' ,data=ratings,palette=['white','red','orange','yellow','green'])
```

## Observation
- Not Rated count is very high
- Maximum number of rating are between 2.5 to 3.4

```python
## count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green'])
```
